Carga de datos (2 pts)

```python
# Se cargan las bibliotecas pertinentes para este ejercicio con Tensorflow, y la base de datos fashion mnist
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import fashion_mnist
import matplotlib.pyplot as plt
```

```python
# Se dividen los datos en conjuntos de entrenamiento y prueba.
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

##**üöÄPreprocesamiento (2 pts)**

##Escala los valores de los p√≠xeles a un rango [0, 1].

##Convierte las etiquetas (y_train, y_test) a codificaci√≥n categ√≥rica (to_categorical).

```python
## Se escalan los datos entre 0 y 1, se divide el test y el train entre 255 que es la escala de grices
x_train = x_train/255.0
x_test = x_test/255.0
# Se imprime una verificaci√≥n de los datos
print(f"Valor m√≠nimo: {x_train.min()}, Valor m√°ximo: {x_train.max()}")
```

```python
# Convierte las etiquetas (y_train, y_test) a codificaci√≥n categ√≥rica (to_categorical).
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)
# Se verifican los resultados
print(f"Forma de y_train: {y_train.shape}")
print(f"Forma de y_test: {y_test.shape}")
```

##**Construcci√≥n de la red CNN (4 pts)**

Incluye al menos:

üéØUna capa Conv2D con activaci√≥n ReLU.

üéØUna capa MaxPooling2D.

üéØUna capa Flatten.

üéØUna capa Dense de salida con activaci√≥n softmax.

üéØMuestra la arquitectura con .summary().

```python
# Se define el modelo.
modelo = Sequential([
    Conv2D(32, (3,3), activation="relu", input_shape=(28,28,1)), # Capa convulocional con activaci√≥n ReLU.
    MaxPooling2D(pool_size=(2,2)), # Capa de Maxpoolin
    Flatten(), # Aplanamiento de la capa densa
    Dense(10, activation="softmax")  # Capa de salida softmax
])

modelo.summary()  # Se muestra el summary del modelo.
```

##**Entrenamiento del modelo (2 pts)**

Usa categorical_crossentropy y la m√©trica accuracy.

Entrena entre 5 y 10 √©pocas.

Visualiza la p√©rdida y precisi√≥n en entrenamiento y validaci√≥n.

```python
# Se usa crossentropy para la perdida, se optimiza con Adam, y se usa la m√©trica Accuracy
modelo.compile(optimizer="adam",
               loss="categorical_crossentropy",
               metrics=["accuracy"]
               )

# Se entrena el modelo en 32 batches.
history = modelo.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test))


import numpy as np

# Encontrar la √©poca √≥ptima (m√≠nima p√©rdida en validaci√≥n)
epoca_optima = np.argmin(history.history['val_loss'])  # √çndice de la mejor √©poca
perdida_optima = history.history['val_loss'][epoca_optima]  # Valor de p√©rdida en esa √©poca

plt.figure(figsize=(12,5))

# üìâ P√©rdida en entrenamiento y validaci√≥n
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Entrenamiento')
plt.plot(history.history['val_loss'], label='Validaci√≥n')
plt.scatter(epoca_optima, perdida_optima, color='red', marker='x', s=100, label="√âpoca √≥ptima")  # Marca de la mejor √©poca
plt.xlabel('√âpocas')
plt.ylabel('P√©rdida')
plt.title('P√©rdida en entrenamiento y validaci√≥n')
plt.legend()

plt.show()
```

##üöÄ Al medir la perdida y las √©pocas se ve que el punto √≥ptimo se encuentra entre la primera y segunda √©poca.  Dado que el m√≥delo llega a un 90% de precisi√≥n y la perdida es la mejor con 0.29, despu√©s de esas √©pocas la perdida aumenta un poco.


## üéØEvaluaci√≥n y an√°lisis (3 pts)

Eval√∫a el modelo sobre el conjunto de prueba.

```python
# Evaluar el rendimiento en el conjunto de prueba
resultado = modelo.evaluate(x_test, y_test)

# Mostrar los resultados
print(f"P√©rdida en prueba: {resultado[0]:.4f}")
print(f"Precisi√≥n en prueba: {resultado[1]*100:.2f}%")
```

##üéØ **Resultados de la evaluaci√≥n:**

‚úÖ La perdida se encuentra en 0.335 en prueba lo cual indica que no existe un sobreajuste del modelo, al igual que no existen una gran perdida de datos.

‚úÖ La precisi√≥n se encuentra en el 90% en prueba lo cual indica un rendimiento muy bueno al clasificar las imagenes de Fasion-MNIST.

‚úÖ La precision en prueba es muy parecida o cercana a la de entrenamiento, lo cual indica que no existen sobreajustes.

##**üß© Reto opcional ‚Äì Mejora con regularizaci√≥n**
üéØAgrega una segunda capa convolucional y aplica Dropout entre capas.

üéØVuelve a entrenar y compara los resultados con el modelo base.

üéØComenta si mejor√≥ o empeor√≥ la generalizaci√≥n.

```python


# Definir el modelo mejorado
modelo_mejorado = Sequential([
    Conv2D(32, (3,3), activation="relu", input_shape=(28,28,1)),  # Primera capa convolucional
    MaxPooling2D(pool_size=(2,2)),  # Primera capa de MaxPooling

    Conv2D(64, (3,3), activation="relu"),  # Segunda capa convolucional
    MaxPooling2D(pool_size=(2,2)),  # Segunda capa de MaxPooling

    Flatten(),  # Aplanamiento de la capa densa

    Dropout(0.5),  # Dropout para reducir sobreajuste

    Dense(128, activation="relu"),  # Capa densa intermedia con m√°s neuronas
    Dropout(0.5),  # Otro Dropout antes de la salida

    Dense(10, activation="softmax")  # Capa de salida softmax
])

# Mostrar el summary del modelo mejorado
modelo_mejorado.summary()
```

##üéØ**Vuelve a entrenar y compara los resultados con el modelo base.**

```python
modelo_mejorado.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
# Cargar el dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Preprocesamiento: Normalizar y cambiar dimensiones
X_train = X_train / 255.0
X_test = X_test / 255.0
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

# Entrenar el modelo mejorado
historia_mejorado = modelo_mejorado.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))
```
