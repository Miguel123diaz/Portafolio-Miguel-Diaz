##🔧**Instrucciones:**
Carga de datos (2 pts)

Utiliza el conjunto fashion_mnist de Keras.

Divide los datos en conjuntos de entrenamiento y prueba.

```python
# Se cargan las bibliotecas pertinentes para este ejercicio con Tensorflow, y la base de datos fashion mnist
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import fashion_mnist
import matplotlib.pyplot as plt
```

```python
# Se dividen los datos en conjuntos de entrenamiento y prueba.
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

##**🚀Preprocesamiento (2 pts)**

##Escala los valores de los píxeles a un rango [0, 1].

##Convierte las etiquetas (y_train, y_test) a codificación categórica (to_categorical).

```python
## Se escalan los datos entre 0 y 1, se divide el test y el train entre 255 que es la escala de grices
x_train = x_train/255.0
x_test = x_test/255.0
# Se imprime una verificación de los datos
print(f"Valor mínimo: {x_train.min()}, Valor máximo: {x_train.max()}")
```

```python
# Convierte las etiquetas (y_train, y_test) a codificación categórica (to_categorical).
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)
# Se verifican los resultados
print(f"Forma de y_train: {y_train.shape}")
print(f"Forma de y_test: {y_test.shape}")
```

##**Construcción de la red CNN (4 pts)**

Incluye al menos:

🎯Una capa Conv2D con activación ReLU.

🎯Una capa MaxPooling2D.

🎯Una capa Flatten.

🎯Una capa Dense de salida con activación softmax.

🎯Muestra la arquitectura con .summary().

```python
# Se define el modelo.
modelo = Sequential([
    Conv2D(32, (3,3), activation="relu", input_shape=(28,28,1)), # Capa convulocional con activación ReLU.
    MaxPooling2D(pool_size=(2,2)), # Capa de Maxpoolin
    Flatten(), # Aplanamiento de la capa densa
    Dense(10, activation="softmax")  # Capa de salida softmax
])

modelo.summary()  # Se muestra el summary del modelo.
```

##**Entrenamiento del modelo (2 pts)**

Usa categorical_crossentropy y la métrica accuracy.

Entrena entre 5 y 10 épocas.

Visualiza la pérdida y precisión en entrenamiento y validación.

```python
# Se usa crossentropy para la perdida, se optimiza con Adam, y se usa la métrica Accuracy
modelo.compile(optimizer="adam",
               loss="categorical_crossentropy",
               metrics=["accuracy"]
               )

# Se entrena el modelo en 32 batches.
history = modelo.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test))


import numpy as np

# Encontrar la época óptima (mínima pérdida en validación)
epoca_optima = np.argmin(history.history['val_loss'])  # Índice de la mejor época
perdida_optima = history.history['val_loss'][epoca_optima]  # Valor de pérdida en esa época

plt.figure(figsize=(12,5))

# 📉 Pérdida en entrenamiento y validación
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Entrenamiento')
plt.plot(history.history['val_loss'], label='Validación')
plt.scatter(epoca_optima, perdida_optima, color='red', marker='x', s=100, label="Época óptima")  # Marca de la mejor época
plt.xlabel('Épocas')
plt.ylabel('Pérdida')
plt.title('Pérdida en entrenamiento y validación')
plt.legend()

plt.show()
```

##🚀 Al medir la perdida y las épocas se ve que el punto óptimo se encuentra entre la primera y segunda época.  Dado que el módelo llega a un 90% de precisión y la perdida es la mejor con 0.29, después de esas épocas la perdida aumenta un poco.


## 🎯Evaluación y análisis (3 pts)

Evalúa el modelo sobre el conjunto de prueba.

```python
# Evaluar el rendimiento en el conjunto de prueba
resultado = modelo.evaluate(x_test, y_test)

# Mostrar los resultados
print(f"Pérdida en prueba: {resultado[0]:.4f}")
print(f"Precisión en prueba: {resultado[1]*100:.2f}%")
```

##🎯 **Resultados de la evaluación:**

✅ La perdida se encuentra en 0.335 en prueba lo cual indica que no existe un sobreajuste del modelo, al igual que no existen una gran perdida de datos.

✅ La precisión se encuentra en el 90% en prueba lo cual indica un rendimiento muy bueno al clasificar las imagenes de Fasion-MNIST.

✅ La precision en prueba es muy parecida o cercana a la de entrenamiento, lo cual indica que no existen sobreajustes.

🎯Comenta si mejoró o empeoró la generalización.

```python


# Definir el modelo mejorado
modelo_mejorado = Sequential([
    Conv2D(32, (3,3), activation="relu", input_shape=(28,28,1)),  # Primera capa convolucional
    MaxPooling2D(pool_size=(2,2)),  # Primera capa de MaxPooling

    Conv2D(64, (3,3), activation="relu"),  # Segunda capa convolucional
    MaxPooling2D(pool_size=(2,2)),  # Segunda capa de MaxPooling

    Flatten(),  # Aplanamiento de la capa densa

    Dropout(0.5),  # Dropout para reducir sobreajuste

    Dense(128, activation="relu"),  # Capa densa intermedia con más neuronas
    Dropout(0.5),  # Otro Dropout antes de la salida

    Dense(10, activation="softmax")  # Capa de salida softmax
])

# Mostrar el summary del modelo mejorado
modelo_mejorado.summary()
```

##🎯**Vuelve a entrenar y compara los resultados con el modelo base.**
